```python
user_model = '''const mongoose = require('mongoose');

const userSchema = new mongoose.Schema({
  userId: {
    type: String,
    unique: true,
    required: true,
    default: () => new mongoose.Types.ObjectId().toString()
  },
  username: {
    type: String,
    unique: true,
    required: true,
    trim: true,
    minlength: 3
  },
  email: {
    type: String,
    unique: true,
    required: true,
    lowercase: true,
    trim: true
  },
  passwordHash: {
    type: String,
    required: true
  },
  profilePic: {
    type: String,
    default: 'https://res.cloudinary.com/demo/image/upload/avatar.png'
  },
  bio: {
    type: String,
    maxlength: 500,
    default: ''
  },
  level: {
    type: Number,
    default: 1
  },
  xp: {
    type: Number,
    default: 0
  },
  friends: [{
    type: mongoose.Schema.Types.ObjectId,
    ref: 'User'
  }],
  location: {
    type: {
      type: String,
      enum: ['Point'],
      default: 'Point'
    },
    coordinates: {
      type: [Number],
      default: [0, 0]
    }
  },
  inventory: [{
    type: mongoose.Schema.Types.ObjectId,
    ref: 'Character'
  }],
  currency: {
    berries: { type: Number, default: 1000 },
    poneglyphs: { type: Number, default: 100 }
  },
  premium: {
    type: Boolean,
    default: false
  },
  pityCounter: {
    type: Number,
    default: 0
  },
  lastLogin: {
    type: Date,
    default: Date.now
  },
  createdAt: {
    type: Date,
    default: Date.now
  }
}, {
  timestamps: true
});

userSchema.index({ location: '2dsphere' });
userSchema.index({ username: 1, email: 1 });

module.exports = mongoose.model('User', userSchema);
'''

with open('PoneglyphPulse/backend/models/User.js', 'w') as f:
    f.write(user_model)

# Character Model  
character_model = '''const mongoose = require('mongoose');

const characterSchema = new mongoose.Schema({
  characterId: {
    type: String,
    unique: true,
    required: true
  },
  name: {
    type: String,
    required: true,
    trim: true
  },
  series: {
    type: String,
    default: 'One Piece'
  },
  faction: {
    type: String,
    enum: ['Straw Hat Pirates', 'Marines', 'Revolutionary Army', 'Yonko', 'Warlords', 'Other'],
    default: 'Other'
  },
  rarity: {
    type: String,
    enum: ['Common', 'Uncommon', 'Rare', 'Epic', 'Legendary', 'Mythic', 'Secret'],
    required: true
  },
  stats: {
    strength: { type: Number, default: 10 },
    speed: { type: Number, default: 10 },
    durability: { type: Number, default: 10 },
    haki: { type: Number, default: 0 },
    devilFruit: { type: Boolean, default: false }
  },
  abilities: [{
    name: String,
    description: String,
    power: Number
  }],
  image: {
    type: String,
    required: true
  },
  tradeable: {
    type: Boolean,
    default: true
  },
  marketValue: {
    type: Number,
    required: true
  },
  description: {
    type: String,
    maxlength: 1000
  },
  dropRate: {
    type: Number,
    required: true
  },
  createdAt: {
    type: Date,
    default: Date.now
  }
}, {
  timestamps: true
});

characterSchema.index({ rarity: 1, name: 1 });
characterSchema.index({ characterId: 1 });

module.exports = mongoose.model('Character', characterSchema);
'''

with open('PoneglyphPulse/backend/models/Character.js', 'w') as f:
    f.write(character_model)

print("✅ Model files created (5/13)")
```
